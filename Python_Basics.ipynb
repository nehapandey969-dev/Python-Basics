Python Basics – Theory Answers
1. What is Python, and why is it popular?
Python is a high-level, interpreted programming language known for its simplicity and readability.
It is popular because:
•	Easy to learn and use
•	Large standard library
•	Supports multiple programming paradigms (procedural, OOP, functional)
•	Platform-independent

2. What is an interpreter in Python?
An interpreter is a program that executes Python code line-by-line, converting it into machine-readable instructions at runtime.
Python uses an interpreter (CPython is the default).

3. What are pre-defined keywords in Python?
Keywords are reserved words in Python that have a predefined meaning and cannot be used as variable names. Examples: if, else, while, True, None.

4. Can keywords be used as variable names?
No, keywords cannot be used as variable names because they are reserved for specific syntax and functions.

5. What is mutability in Python?
Mutability means whether an object’s content can be changed after creation.
•	Mutable objects: Can be modified (e.g., list, set).
•	Immutable objects: Cannot be modified (e.g., tuple, string, int)
6. Why are lists mutable, but tuples are immutable?
A list is a mutable, ordered sequence of items. This means that a list is created its elements can be modified. (added,  removed or changed.)
A tuple is an immutable , ordered  sequence of item. This means that once a tuple is created its elements cannot be changed.

7. Difference between == and is operators in Python
•	== → Checks value equality (whether the values are the same).
•	is → Checks identity (whether two references point to the same object in memory).

8. What are logical operators in Python?
Logical operators combine conditional statements:
•	and – True if both conditions are true
•	or – True if at least one condition is true
•	not – Reverses the condition’s result
9. What is type casting in Python?
Type casting is converting a variable from one data type to another (e.g., int → float).

10. Difference between implicit and explicit type casting
•	Implicit (Type Conversion): this occurs automatically by the python compiler.
•	Explicit (Type Casting): It is performed by the programmer using build in function.

11. Purpose of conditional statements in Python
They are used to execute specific code only when certain conditions are met.
12. How does the elif statement work?
elif  statement in python is a shorthand for “else if”. It is used as part of a conditional statement.
13. Difference between for and while loops
•	for loop → Used when the number of iterations is known.
•	while loop → Used when iterations depend on a condition being true.

14. Scenario where a while loop is more suitable than a for loop
When taking input from a user until they type “exit” — because the number of iterations is unknown.
